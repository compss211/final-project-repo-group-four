```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing cleand and processed data
df = pd.read_csv('../data/clean_processed.csv')
df.head()
```

```python
# Creating a Parent Vocalization List 
vocalizations = ["ah","aah","aaah","aaaaah","aaaahh","aaaahhh","aagh","aww","awww","ayy","ayyy","bah","bam","bap","brr",
                 "choo","da","daa","dah","dee","doo","doot","dum","fa","goo","hah","haaa","hahh","ha-ha","ha-ha-ha","hahhh",
                 "haha","heh","hey","ho","hoh","hoo","hmm","hmmm","hmmmm","la","laa","lalala","la-la-la","lah","mamama",
                 "mm","mmm","mmmm","nah","naah","naaah","nahh","o","oh","ohh","ohhh","ohhhh","ooh","oooh","ooooh","ow","oww",
                 "ra","rah","sha","shaa","shh","shhh","ta","taa","uh","uhh","uhhh","uh-huh","unh","wah","waaah","whoa","whoaa",
                 "whoah","woo","wooo","woooo","yeah","yeahh","yeahhh","yo","yoo","yooo","yoooo","yum"]
```

```python
# Splitting Corpus into individual tokens to compare vocalizations with 
df['tokens'] = df['Corpus'].str.split()
```

```python
# Doing the comparisons against vocalizations in corpus 
vocal_set = set(vocalizations)

df['vocalization_count'] = df['tokens'].apply(lambda toks: sum(t in vocal_set for t in toks))
```

```python
# Plotting the average vocalization counts across years 
df.groupby('Year')['vocalization_count'].mean().plot()
plt.xlabel("Year")
plt.ylabel("Average vocalizations per song")
plt.title("Vocalization Density Over Time")
plt.show()
```
