```python
df = pd.read_csv("data/clean_processed.csv")
df.head()
# we work on Lyrics, full sentence
```

```python
# separte into different df based on decades
df_1960s= df[(df["Year"]>= 1960) & (df["Year"] < 1970)]
print("Number of songs in 1960s:", len(df_1960s))
df_1970s= df[(df["Year"]>= 1970) & (df["Year"] < 1980)]
print("Number of songs in 1970s:", len(df_1970s))
df_1980s= df[(df["Year"]>= 1980) & (df["Year"] < 1990)]
print("Number of songs in 1980s:", len(df_1980s))
df_1990s= df[(df["Year"]>= 1990) & (df["Year"] < 2000)]
print("Number of songs in 1990s:", len(df_1990s))
df_2000s= df[(df["Year"]>= 2000) & (df["Year"] < 2010)]
print("Number of songs in 2000s:", len(df_2000s))
df_2010s= df[(df["Year"]>= 2010) & (df["Year"] < 2020)]
print("Number of songs in 2010s:", len(df_2010s))
```

```python
!pip install vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()
```

### 1960s

```python
df_1960s['sentiment_scores'] = df_1960s['Lyrics'].apply(
     lambda x: analyzer.polarity_scores(str(x))
)

# Extract compound score (overall sentiment)
df_1960s['compound'] = df_1960s['sentiment_scores'].apply(lambda x: x['compound'])
df_1960s['pos'] = df_1960s['sentiment_scores'].apply(lambda x: x['pos'])
df_1960s['neg'] = df_1960s['sentiment_scores'].apply(lambda x: x['neg'])
df_1960s['neu'] = df_1960s['sentiment_scores'].apply(lambda x: x['neu'])

df_1960s.head()
```

```python
def categorize_sentiment(compound):
    """
    Categorize sentiment based on compound score.
    Adjust thresholds based on your data if needed.
    """
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# Apply categorization
df_1960s['sentiment_category'] = df_1960s['compound'].apply(categorize_sentiment)

# Show distribution
print("Sentiment distribution:")
print(df_1960s['sentiment_category'].value_counts())
print("\nPercentages:")
print(df_1960s['sentiment_category'].value_counts(normalize=True).round(3) * 100)
```

### 1960s - 2010s

```python
# 1960s2010s sentiment analysis
# pip install vaderSentiment matplotlib pandas

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import pandas as pd
import matplotlib.pyplot as plt

# initialize VADER
analyzer = SentimentIntensityAnalyzer()

# all the decades into a dict
decades = {
    '1960s': df_1960s,
    '1970s': df_1970s,
    '1980s': df_1980s,
    '1990s': df_1990s,
    '2000s': df_2000s,
    '2010s': df_2010s
}

# define the category
def categorize_sentiment(compound):
    if compound >= 0.05:
        return 'positive'
    elif compound <= -0.05:
        return 'negative'
    else:
        return 'neutral'

# run VADER for each decades
for name, df in decades.items():
    df = df.copy()
    df['Lyrics'] = df['Lyrics'].fillna('').astype(str)
    df['sentiment_scores'] = df['Lyrics'].apply(lambda x: analyzer.polarity_scores(x))
    df['compound'] = df['sentiment_scores'].apply(lambda x: x['compound'])
    df['sentiment_category'] = df['compound'].apply(categorize_sentiment)
    decades[name] = df  # update the DataFrame

# combine all the decades
df_all = pd.concat(decades.values(), ignore_index=True)
df_all = df_all[df_all['Year'].notna()].copy()
df_all['Year'] = df_all['Year'].astype(int)

# Use years for new segmentation
# ========= calculate each years' sentiment category's proportion =========
yearly_counts = df_all.groupby(['Year', 'sentiment_category']).size().unstack(fill_value=0)
yearly_percent = yearly_counts.div(yearly_counts.sum(axis=1), axis=0)

# ========= Plotting =========
plt.figure(figsize=(10,6))
plt.plot(yearly_percent.index, yearly_percent['positive'], label='Positive', color='#4C72B0', linewidth=2)
plt.plot(yearly_percent.index, yearly_percent['neutral'], label='Neutral', color='#55A868', linewidth=2)
plt.plot(yearly_percent.index, yearly_percent['negative'], label='Negative', color='#C44E52', linewidth=2)

plt.title("Sentiment Trends in Lyrics (1960s2010s)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Proportion of Sentiment Category")
plt.ylim(0, 1)
plt.grid(alpha=0.3)
plt.legend(title="Sentiment", fontsize=10)
plt.tight_layout()
plt.show()
```
